```python
# make a barplot for chr_accs
import numpy as np


plt.figure(figsize=figsize)
plt.bar(df['chr'], df['val_acc'])

# Add a horizontal line for the average accuracy
avg_acc = np.mean(df['val_acc'])
plt.axhline(avg_acc, color='red', linestyle='--', linewidth=1.5, label=f'Average: {avg_acc:.1f}%')

plt.xlabel('Chromosome')
plt.ylabel('Accuracy (%)')
plt.title('Whole-genome pretrain')
plt.tight_layout()
plt.ylim(95,99)
plt.legend()
plt.savefig('whole-genome_pretrain_performance.pdf')
plt.show()
plt.close()
```
