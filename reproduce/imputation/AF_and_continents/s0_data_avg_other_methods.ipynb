```python
indir = './s0_result'
outdir = indir
```

```python
import pandas as pd
```

```python
all_df = []
phasings = ['beagle', 'shapeit', 'eagle']
imputations = ['beagle', 'minimac', 'impute']
refs = ['noREF', 'REF']
for phasing in phasings:
    for imputation in imputations:
        for ref in refs:
            method = f'{phasing}_{ref}_{imputation}'
            fn = f'{indir}/{method}.tsv'
            df = pd.read_csv(fn, sep='\t')
            df['ref'] = ref
            all_df.append(df)

all_df = pd.concat(all_df)
all_df.head()
```

```python
# group by sample, maf_bin and ref
avg_df = all_df.groupby(['ref', 'continent', 'sample', 'maf_bin'])['accuracy'].mean().reset_index()
avg_df.head()

# subset to only include noREF
noREF_df = avg_df[avg_df['ref'] == 'noREF']
noREF_df.to_csv(f'{outdir}/noREF.tsv', sep='\t', index=False)

# subset to only include REF
REF_df = avg_df[avg_df['ref'] == 'REF']
REF_df.to_csv(f'{outdir}/REF.tsv', sep='\t', index=False)
```
